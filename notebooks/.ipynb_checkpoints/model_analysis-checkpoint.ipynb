### Visual comparisons of the three baselines
**RMSE**, **MAE**, and $ R^2 $ scores are used.

$R^2$ is a statistical metric that explains the proportion of the variance in the dependent variable that is predictable from the independent variables.

It is defined as:

$$
R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}
$$

Where:
- $y_i$ is the actual value for the $i$-th observation,
- $\hat{y}_i$ is the predicted value for the $i$-th observation,
- $\bar{y}$ is the mean of the actual values, and
- $n$ is the total number of observations.

The $R^2$ score ranges from 0 to 1, where:
- $R^2 = 1$ indicates a perfect fit, and
- $R^2 = 0$ indicates that the model does not explain any of the variance in the data.

# Section 3: Evaluating Feature Selection

The top `k` features that have the highest $F$ scores are kept, where the $F$ scores are computed **individually for each feature $x^{(j)}$** as:

$$
F^{(j)} = \frac{\frac{1}{n - 1} \sum_{i=1}^{n} \left( \hat{y}_i^{(j)} - \bar{y} \right)^2}{\frac{1}{n} \sum_{i=1}^{n} \left( y_i - \hat{y}_i^{(j)} \right)^2}
$$

Where:
- $F^{(j)}$ is the F-score for the $j$-th feature,
- $y_i$ is the true target value for the $i$-th sample,
- $\hat{y}_i^{(j)}$ is the predicted value of $y_i$ using only the $j$-th feature (using a simple univariate linear regression model),
- $\bar{y}$ is the mean of the target values: $\bar{y} = \frac{1}{n} \sum_{i=1}^n y_i$,
- $n$ is the total number of samples.

This score measures how well feature $x^{(j)}$ alone explains the variance in the target variable. High $F$ values indicate strong predictive power for that feature.
